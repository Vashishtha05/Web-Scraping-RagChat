```python
from urllib.parse import urlparse
import requests
from langchain_core.documents import Document
```

```python
file_path = 'pages_name.txt'  


lists=[]
with open(file_path, 'r') as file:
    for line in file:
        url = line.strip()  
        lists.append(url)
```

```python
lists
```

```python
from bs4 import BeautifulSoup
import re
def parse_html(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    return soup.get_text()

def fetch_webpage(url):
    response = requests.get(url)
    return response.text

def clean_text(text):
    # Remove extra whitespace and newlines
    text = ' '.join(text.split())
    # Remove any remaining unwanted characters or patterns (example: URLs, special characters)
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'[^a-zA-Z0-9\s.,!?\'"]+', '', text)  # Remove special characters except punctuation
    return text

from docx import Document

def write_to_docx(text_content, output_filename):
    doc = Document()  # Create a new Document object

    # Add a heading
    doc.add_heading('Document Title', level=1)

    doc.add_paragraph(text_content)

    # Save the document
    doc.save(output_filename)
```

```python
for url in lists:
        html_content=fetch_webpage(url)
        page_text=parse_html(html_content)
        final_text=clean_text(page_text)
        output_filename=url.split('/')[-1]+'.docx'

        write_to_docx(final_text,output_filename)
```

```python
from docx2pdf import convert
import os

# Current folder = scraping
input_folder = os.getcwd()

# Go to project root and create static folder there
BASE_DIR = os.path.dirname(input_folder)
output_folder = os.path.join(BASE_DIR, "static")

os.makedirs(output_folder, exist_ok=True)

for file in os.listdir(input_folder):
    if file.lower().endswith(".docx"):
        docx_path = os.path.join(input_folder, file)
        pdf_path = os.path.join(output_folder, file.replace(".docx", ".pdf"))

        print("Converting:", docx_path)
        convert(docx_path, pdf_path)
```
